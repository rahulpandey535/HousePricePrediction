```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
#importing boston dataset
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
```

```python
print(housing)
```

```python
#loading data in pandas data frame 
house_data=pd.DataFrame(housing.data,columns=housing.feature_names)
```

```python
house_data.head()
```

```python
#add the target column in to the data frame 
house_data['Price']=housing.target
```

```python
house_data.head()
```

```python
#checking the number of rows and columns in the data frame 
house_data.shape
```

```python
#check for missing value
house_data.isnull().sum()
```

```python
#statistical analysis 
house_data.describe()
```

```python
#understanding the co relation between various features int the daata set
#positive and negative corelation  
correlation=house_data.corr()
```

```python
#constructing a heat map to understand the correltion'
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Blues')
```

```python
#splitting the data and target
x=house_data.drop(['Price'],axis=1)
y=house_data['Price']
```

```python
print(x)
```

```python
print(y)
```

```python
#splitting the data into training data and test data 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
print(y.shape,y_train.shape,y_test.shape)
```

```python
#model training 
model=XGBRegressor()
```

```python
#fitting model by x train
model.fit(x_train,y_train)
```

```python
#model evaluation
#prediction on training data
#accuracy for predicton on traing data
training_data_prediction=model.predict(x_train)
```

```python
print(training_data_prediction)
```

```python
#R square error

score_1=metrics.r2_score(y_train,training_data_prediction)

#mean absolute error

score_2=metrics.mean_absolute_error(y_train,training_data_prediction)

print("R Squre error is :",score_1)

print("Mean absolute error is :",score_2)

```

```python
#vizulising 
plt.scatter(y_train,training_data_prediction)
plt.xlabel("Actuyal Price")
plt.ylabel("Predected Price")
plt.title("Actual Price Vs predicted Price")
plt.show()
```

```python
#predicton on test data 
test_data_prediction=model.predict(x_test)
```

```python
print(test_data_prediction)
```

```python
#R square error

score_1=metrics.r2_score(y_test,test_data_prediction)

#mean absolute error

score_2=metrics.mean_absolute_error(y_test,test_data_prediction)

print("R Squre error is :",score_1)

print("Mean absolute error is :",score_2)
```

```python
#vizulising 
plt.scatter(y_test,test_data_prediction)
plt.xlabel("Actuyal Price")
plt.ylabel("Predected Price")
plt.title("Actual Price Vs predicted Price")
plt.show()
```
